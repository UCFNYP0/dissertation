```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('destination.csv')
```

```python
columns= ['index','station_id','x','y','docks_num','centrality','centrali_1']
df1=df[columns]
```

```python
df1
```

```python
dis = pd.read_csv('distance matrix.csv')
```

```python
dis
```

```python
dis_filtered = dis.copy()
dis_filtered = dis_filtered.drop(columns=["ID"])
#filter less than 5km
dis_filtered = dis_filtered.applymap(lambda x: x if x < 5000 else pd.NA)
```

```python
dis_filtered
```

```python
# 把 docks_num 列设置为索引，方便之后的运算
df1.set_index('station_id', inplace=True)
```

```python
df1.index = df1.index.astype(str)
```

```python
dis_filtered = dis_filtered.fillna(0).astype(float)
dis_filtered.replace(0, np.nan, inplace=True)


# β
beta = 2

df_AD = pd.DataFrame(index=dis_filtered.columns)

for column in dis_filtered.columns:
    df_AD.loc[column, 'AD'] = np.nansum(df1.loc[column, 'docks_num'] * (dis_filtered[column] ** -beta))
```

```python
df1
```

```python
df_AD
```

```python
merged_AD = df1.merge(df_AD, left_index=True, right_index=True)
```

```python
merged_AD
```

```python
merged_AD.to_csv('station destination.csv')
```
